```python
import numpy as np
import pandas as pd
import matplotlib 
import seaborn as sns

```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding="ISO-8859-1")
```

```python
df.columns
```

```python
df.dropna(axis=1, how='all', inplace=True)
```

```python
pd.isnull(df).sum()  # check no ofnull values in clms 
```

```python
df.rename(columns={'Marital_Status':'Marital Status'}, inplace=True)
```

```python
df.describe()
```

```python
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum()
```

```python
gax = sns.barplot(x='Gender', y ='Amount', data=sales_by_gender)
for bars in gax.containers:
    gax.bar_label(bars)
```
